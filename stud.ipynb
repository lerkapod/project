```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
from PIL import ImageTk,Image
import io
import sqlite3


# главное окно
class Main(tk.Frame):
    def __init__(self, root):
        super().__init__(root)
        self.init_main()
        self.db = db
        self.view_records()

    def init_main(self):
        toolbar = tk.Frame(bg='#B0E0E6', bd=2)
        toolbar.pack(side=tk.TOP, fill=tk.X)

        self.add_img = tk.PhotoImage(file='add-stud.png')
        btn_open_dialog = tk.Button(toolbar, text='Добавить позицию', command=self.open_dialog, bg='#B0E0E6', bd=0,
                                    compound=tk.TOP, image=self.add_img)
        btn_open_dialog.pack(side=tk.LEFT)

        self.edit_img = tk.PhotoImage(file='upd.png')
        btn_edit_dialog = tk.Button(toolbar, text='Редактировать', bg='#B0E0E6', bd=0,
                                    compound=tk.TOP, command=self.open_update_dialog, image=self.edit_img)
        btn_edit_dialog.pack(side=tk.LEFT)

        self.delete_img = tk.PhotoImage(file='delete.png')
        btn_delete = tk.Button(toolbar, text='Удалить позицию', bg='#B0E0E6', bd=0,
                               compound=tk.TOP, command=self.delete_records, image=self.delete_img)
        btn_delete.pack(side=tk.LEFT)

        self.gal_img = tk.PhotoImage(file='gallery.png')
        btn_1 = tk.Button(toolbar, text='Посмотреть фото', bg='#B0E0E6', bd=0, 
                               compound=tk.TOP, command=self.open_dialog1, image=self.gal_img)
        btn_1.pack(side=tk.LEFT)
        
        self.tree = ttk.Treeview(self,
                                 columns=('id_stud', 'name', 'year', 'average_ball', 'photo', 'course', 'id_group'),
                                 height=25, show='headings')

        self.tree.column('id_stud', width=30, anchor=tk.CENTER)
        self.tree.column('name', width=100, anchor=tk.CENTER)
        self.tree.column('year', width=100, anchor=tk.CENTER)
        self.tree.column('average_ball', width=100, anchor=tk.CENTER)
        self.tree.column('photo', width=100, anchor=tk.CENTER)

        self.tree.column('course', width=100, anchor=tk.CENTER)
        self.tree.column('id_group', width=100, anchor=tk.CENTER)

        self.tree.heading('id_stud', text='ID')
        self.tree.heading('name', text='Имя студента')
        self.tree.heading('year', text='Дата рождения')
        self.tree.heading('average_ball', text='Средний балл')
        self.tree.heading('photo', text='Фото')
        self.tree.heading('course', text='Номер курса')
        self.tree.heading('id_group', text='Номер группы')

        self.tree.pack()

    def records(self, name, year, average_ball, photo, course, id_group):
        self.db.insert_data(name, year, average_ball, photo, course, id_group)
        self.view_records()

    def update_record(self, name, year, average_ball, photo, course, id_group):
        photo = self.db.convertToBinaryData(photo)
        self.db.c.execute(
            '''UPDATE students SET name=?,year=?,average_ball=?,photo=?,course=?,id_group=? WHERE id_stud=?''',
            (name, year, average_ball, photo, course, id_group, self.tree.set(self.tree.selection()[0], '#1')))
        self.db.conn.commit()
        self.view_records()

    def view_records(self):
        self.db.c.execute('''SELECT * FROM students''')
        [self.tree.delete(i) for i in self.tree.get_children()]
        [self.tree.insert('', 'end', values=row) for row in self.db.c.fetchall()]
      
    def delete_records(self):
        for selection_item in self.tree.selection():
            self.db.c.execute('''DELETE FROM students WHERE id_stud=?''', (self.tree.set(selection_item, '#1')))
        self.db.conn.commit()
        self.view_records()

    def open_dialog(self):
        Child()

    def open_update_dialog(self):
        Update()

    def open_dialog1(self):
        Foto()
        
        
# окно добавления
class Child(tk.Toplevel):
    def __init__(self):
        super().__init__(root)
        self.init_child()
        self.view = app

    def show_message(self, txt):
        op = filedialog.askopenfilename()
        self.label_txt["text"] = op

    def init_child(self):
        self.title('Добавить студента')
        self.geometry('400x350+1000+300')
        self.configure(bg='#B0E0E6')
        self.resizable(False, False)

        label_name = tk.Label(self, text='Имя студента:')
        label_name.place(x=50, y=40)
        label_year = tk.Label(self, text='Дата рождения:')
        label_year.place(x=50, y=80)
        label_ball = tk.Label(self, text='Балл:')
        label_ball.place(x=50, y=120)
        label_photo = tk.Label(self, text='Фото:')
        label_photo.place(x=50, y=160)
        label_course = tk.Label(self, text='Номер курса:')
        label_course.place(x=50, y=220)
        label_id_group = tk.Label(self, text='Номер группы:')
        label_id_group.place(x=50, y=260)

        self.entry_name = ttk.Entry(self)
        self.entry_name.place(x=200, y=40)

        arr_year = [i for i in range(1980, 2007)]
        self.cmbx_y = ttk.Combobox(self, values=arr_year, width=10)
        self.cmbx_y.current(20)
        self.cmbx_y.place(x=300, y=80)

        arr_data = [i for i in range(1, 32)]
        self.cmbx_d = ttk.Combobox(self, values=arr_data, width=5)
        self.cmbx_d.current(0)
        self.cmbx_d.place(x=160, y=80)

        arr_m = [i for i in range(1, 13)]
        self.cmbx_m = ttk.Combobox(self, values=arr_m, width=5)
        self.cmbx_m.current(0)
        self.cmbx_m.place(x=230, y=80)

        self.entry_ball = ttk.Entry(self)
        self.entry_ball.place(x=200, y=120)

        self.cmbx_c = ttk.Combobox(self, values=[1, 2, 3, 4, 5])
        self.cmbx_c.current(0)
        self.cmbx_c.place(x=200, y=220)

        self.entry_group = ttk.Entry(self)
        self.entry_group.place(x=200, y=260)

        self.label_txt = tk.Label(self, text='', bg='#B0E0E6')
        self.label_txt.place(x=100, y=180)
        self.b = ttk.Button(self, text='Открыть')
        self.b.place(x=200, y=160)
        self.b.bind('<Button-1>', lambda event: self.show_message(self.label_txt))

        btn_cancel = ttk.Button(self, text='Закрыть', command=self.destroy)
        btn_cancel.place(x=300, y=300)

        self.btn_ok = ttk.Button(self, text='Добавить')
        
        self.btn_ok.place(x=220, y=300)
        self.btn_ok.bind('<Button-1>', lambda event: self.view.records(self.entry_name.get(),
                                                                       self.cmbx_d.get() + '-' +
                                                                       self.cmbx_m.get() + '-' + self.cmbx_y.get(),
                                                                       self.entry_ball.get(),
                                                                       self.label_txt.cget("text"),
                                                                       self.cmbx_c.get(),
                                                                       self.entry_group.get()))

        self.grab_set()  # перехватывает все события
        self.focus_set()  # захватывает и удерживает фокус на текущем окне

# окно редактирования
class Update(Child):
    def __init__(self):
        super().__init__()
        self.init_edit()
        self.view = app

    def init_edit(self):
        self.title('Редактировать позицию')
        btn_edit = ttk.Button(self, text='Редактировать')
        btn_edit.place(x=190, y=300)
        btn_edit.bind('<Button-1>', lambda event: self.view.update_record(self.entry_name.get(),
                                                                          self.cmbx_d.get() + '-' +
                                                                          self.cmbx_m.get() + '-' + self.cmbx_y.get(),
                                                                          self.entry_ball.get(),
                                                                          self.label_txt.cget("text"),
                                                                          self.cmbx_c.get(),
                                                                          self.entry_group.get()))

        self.btn_ok.destroy()
        
# окно просмотра фото    
class Foto(tk.Toplevel):
    def __init__(self):
        super().__init__()
        self.init_foto()
        self.view = app
        
    def selected(self,results,label1):
        name = self.cmbx.get()
        for x in results:
            if x[0]==name:
                foto = x[1]
                img = io.BytesIO(foto)
                image1 = ImageTk.PhotoImage(Image.open(img))
                break
        self.label1["image"] = image1
        self.label1.image = image1
    
    def init_foto(self):
        self.title('Фотография студента')
        self.geometry('600x650+1000+200')
        
                # подпись
        self.label1 = tk.Label(self, text="Здесь будет фотография студента", font=("Arial", 18))  
        self.label1.place(x=10, y=120) 

      
        self.label2 = tk.Label(self, text="Выберите имя студента:", font=("Arial", 14))  
        self.label2.place(x=10, y=30)
        self.cmbx = ttk.Combobox(self, width=50)
        self.cmbx.place(x=250, y=35) 

        #заносим фамилии в combobox
        self.cmbx['values'] = db.view_f()
        self.cmbx.current(1)  # вариант по умолчанию  
       
        self.res = db.view_f1()
        # кнопка button
        self.knopka1 =ttk. Button(self, text="Посмотреть фото")
        self.knopka1.bind('<Button-1>', lambda event: self.selected(self.res,self.label1))
        self.knopka1.place(x=10, y=80) 

# класс с баззой данных
class DB:
    def __init__(self):
        self.conn = sqlite3.connect('bd_Stud.sqlite')
        self.c = self.conn.cursor()
        self.c.execute("""
            CREATE TABLE IF NOT EXISTS students(
            id_stud INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            year TEXT,
            average_ball REAL,
            photo BLOB,
            course INTEGER,
            id_group INTEGER NOT NULL
            );
            """)
        self.conn.commit()

    def convertToBinaryData(self, filename):
        # Преобразование картинки в бинарный формат
        self.blob = open(filename, 'rb')
        self.b = self.blob.read()
        return self.b

    def insert_data(self, name, year, average_ball, photo, course, id_group):
        photo = self.convertToBinaryData(photo)
        self.c.execute("INSERT INTO students(name,year,average_ball,photo,course,id_group)  VALUES (?,?,?,?,?,?)",
                       (name, year, average_ball, photo, course, id_group))
        self.conn.commit()
    def view_f(self):
        #Получаем Фамилии студентов и баллы
        self.c.execute("SELECT name, photo FROM students")
        results = self.c.fetchall()
        names_stud = [x[0] for x in results]
        
        return names_stud
    def view_f1(self):
        #Получаем Фамилии студентов и баллы
        self.c.execute("SELECT name, photo FROM students")
        results = self.c.fetchall()
        names_stud = [x[0] for x in results]
        
        return results

if __name__ == "__main__":
    root = tk.Tk()
    db = DB()
    app = Main(root)
    app.pack()
    root.title("Students")
    root.geometry("650x650+300+200")
    root.resizable(False, False)
    root.mainloop()
```
